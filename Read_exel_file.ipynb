# Print all data with total greather then 200

```python
for i in range(len(alldata)):
    if alldata["TOTAL"][i]>200:
        print(alldata.loc[i])
        print("\n")
```

```python
category=[]
for i in range(0,len(alldata)):
    if alldata["PERCENTAGE"][i]>80 and alldata["PERCENTAGE"][i]<100:
        category.append("SCHOLER")
        
    if alldata["PERCENTAGE"][i]>50 and alldata["PERCENTAGE"][i]<79:
        category.append("AVERAGE")
        
    if alldata["PERCENTAGE"][i]<=50:
        category.append("WEAK")
```

```python
alldata["CATEGORY"]=category
```
