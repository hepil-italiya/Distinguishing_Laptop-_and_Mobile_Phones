```python
import requests
```

```python
url=requests.get("https://data.covid19india.org/data.json")
data=url.json()
```

```python
for i in data:
    print(i)
```

```python
all_date=[]
for i in range(0,len(data["cases_time_series"])):
    if int(data["cases_time_series"][i]["dailyconfirmed"])>10000:
        all_date.append(data["cases_time_series"][i]["date"])
```

```python
count=0
for j in range(0,len(data["cases_time_series"])):
    if count<int(data["cases_time_series"][j]["dailyconfirmed"]):
        count=int(data["cases_time_series"][j]["dailyconfirmed"])
print("Highest case =",count)
```

```python
for p in range(0, len(data["cases_time_series"])):
    if int(data["cases_time_series"][p]["dailyconfirmed"])==count:
        print(data["cases_time_series"][p]["date"])
```

```python
print("Print all date having greterthen 1L cases=")
for n in range(0, len(data["cases_time_series"])):
    if int(data["cases_time_series"][n]["dailyconfirmed"])>100000:
        print(data["cases_time_series"][n]["date"])
```

```python
date=input("Enter date=")
for m in range(0, len(data["cases_time_series"])):
    if data["cases_time_series"][m]["date"]==date:
        if int(data["cases_time_series"][m]["dailyconfirmed"])>100000:
            print("YES")
            break
else:
    print("NO")
```

# MULTIPLE LINE GRAPH

```python
newcase=[]
recover=[]
death=[]
for st1 in range(0,len(data["cases_time_series"])):
    newcase.append(data["cases_time_series"][st1]["dailyconfirmed"])
    recover.append(data["cases_time_series"][st1]["dailyrecovered"])
    death.append(data["cases_time_series"][st1]["dailydeceased"])
```

```python
plt.plot(newcase, color="blue")
plt.plot(recover,color="green")
plt.plot(death,color="red")
plt.show()
```

# START DATE AND END DATE

```python
start=input("ENTER START DATE=")
end=input("ENTER END DATE=")
startlist=[]
startindex=0
endindex=0
for dates in range(0,len(data["cases_time_series"])):
    if start==data["cases_time_series"][dates]["date"]:
        startindex=dates
        
    if end==data["cases_time_series"][dates]["date"]:
        endindex=dates
if endindex>startindex:
    for st2 in range(startindex, endindex):
        startlist.append(data["cases_time_series"][st2]["dailyconfirmed"])
else:
    print("ENTER VALID DATE......")
plt.plot(startlist,color="red")
plt.show()
```
