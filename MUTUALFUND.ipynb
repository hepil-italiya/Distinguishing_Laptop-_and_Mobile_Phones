```python
import requests
```

```python
url=requests.get("https://api.mfapi.in/mf")
data=url.json()
```

```python
count=0
for i in data:
    count=count+1
print("Total NUMBER OF MUTUAL FUND IS = ",count)
```

```python
for j in data:
    print("SCHEMECODE : "+str(j["schemeCode"])+" "+ "SCHEMENAME : "+j["schemeName"])
```

```python
schemecode=int(input("Enter SchemeCode= "))
for m in data:
    if m["schemeCode"]==schemecode:
        print(m["schemeName"])
        break
else:
    print(str(schemecode)+" is not found......")
```
